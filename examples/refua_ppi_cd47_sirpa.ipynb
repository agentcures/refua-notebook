# Refua PPI Notebook: CD47:SIRPalpha Immune Checkpoint Complex

> **Audience:** beginner to advanced.  
> **Goal:** model a biologically important **protein-protein interaction (PPI)** using `Complex([Protein, Protein])`, then run a fold and interpret the setup.

## Why this PPI is interesting
- **Biology:** CD47 on target cells binds SIRPalpha on myeloid cells to transmit an anti-phagocytic checkpoint signal (often called a "don't-eat-me" axis).
- **Drug discovery relevance:** disrupting CD47:SIRPalpha is an active immuno-oncology strategy, including antibody and engineered protein approaches.
- **Structural precedent:** PDB **2JJS** contains the extracellular-domain complex, giving a concrete, literature-backed starting point for a PPI notebook.

## What you will do
1. Define two protein chains from the 2JJS structural system.
2. Build one Refua `Complex([sirpa, cd47])`.
3. Run `fold()` and inspect the resulting complex widget and metadata summary.

## Sequence note
The sequences below come from the RCSB FASTA entry for 2JJS, with terminal expression His-tags removed for cleaner modeling inputs.

```python
from refua import Complex, Protein
```

## Biological setup (plain language)
We model the **Ig-like extracellular domains** that directly form the CD47:SIRPalpha interface in structure 2JJS.

- Chain A/B entity in 2JJS: SIRPalpha domain used for receptor-side recognition.
- Chain C/D entity in 2JJS: CD47 extracellular domain presented by target cells.

This is intentionally a minimal two-protein system to demonstrate core PPI workflow behavior in Refua.


```python
# 2JJS FASTA-derived chain sequences (terminal HHHHHH tag removed).
SIRPA_IG_DOMAIN_2JJS = (
    "EEELQVIQPDKSVSVAAGESAILHCTVTSLIPVGPIQWFRGAGPARELIYNQKEGHFPRVTTVSESTK"
    "RENMDFSISISNITPADAGTYYCVKFRKGSPDTEFKSGAGTELSVRAKPSTR"
)

CD47_IG_DOMAIN_2JJS = (
    "QLLFNKTKSVEFTFGNDTVVIPCFVTNMEAQNTTEVYVKWKFKGRDIYTFDGALNKSTVPTDFSSAKI"
    "EVSQLLKGDASLKMDKSDAVSHTGNYTCEVTELTREGETIIELKYRVVSWSTR"
)

print("SIRPalpha length:", len(SIRPA_IG_DOMAIN_2JJS))
print("CD47 length:", len(CD47_IG_DOMAIN_2JJS))
```

## Build a two-protein Refua complex
For pure PPI systems, `Complex([protein_a, protein_b])` is enough.

Unlike protein-small-molecule setups, there is no ligand-specific `request_affinity(...)` step here; this notebook focuses on **protein-protein structural hypothesis generation**.


```python
sirpa = Protein(SIRPA_IG_DOMAIN_2JJS, ids="A")
cd47 = Protein(CD47_IG_DOMAIN_2JJS, ids="B")

ppi_spec = Complex([sirpa, cd47], name="cd47_sirpa_ppi")
ppi_spec
```

```python
study_context = {
    "axis": "CD47:SIRPalpha innate immune checkpoint",
    "system_type": "protein-protein interaction",
    "entities": len(ppi_spec.entities),
    "entity_ids": [entity.ids for entity in ppi_spec.entities],
    "sequence_lengths": [len(entity.sequence) for entity in ppi_spec.entities],
}
study_context
```

## Interpretation guide before running
`fold()` will generate a structure hypothesis for this two-chain complex.

When reading results, focus on:
- Whether the two domains dock in a plausible interface orientation.
- Whether chain-level geometry is compatible with known checkpoint biology.
- How this baseline could support next steps (mutational scans, binder design, or alternative receptor variants).


## Run the fold
This executes the Refua folding pipeline for the PPI system.


```python
result = ppi_spec.fold()
result
```

```python
# The complex widget now reflects the latest fold result.
ppi_spec
```
